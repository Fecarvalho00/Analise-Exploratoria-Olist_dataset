## Juntando tudo

Este trecho de cdigo combina os DataFrames `orders_df`, `customers_df`, `orderitem_df`, `products_df`, `categname_df`, `orderpay_df`, `sellers_df` e `orderreviews_df`, usando o mtodo `merge()` da biblioteca `pandas`. Cada DataFrame  combinado com o prximo usando uma coluna em comum como chave de juno. 

O resultado da combinao  salvo em um novo DataFrame chamado `dados_completos`. Ao final dessa operao, o DataFrame `dados_completos` contm informaes sobre os pedidos, clientes, itens do pedido, produtos, categorias de produtos, pagamentos, vendedores e avaliaes de pedidos.

Esse processo de combinar vrios DataFrames em um s  til para consolidar dados dispersos em um nico lugar e facilitar a anlise dos dados.

```python
df = orders_df.merge(customers_df, on = 'customer_id').merge(orderitem_df, on = 'order_id').merge(products_df, on = 'product_id').merge(categname_df, on = 'product_category_name').merge(orderpay_df, on = 'order_id').merge(sellers_df, on = 'seller_id').merge(orderreviews_df, on='order_id')
```

Anlise dos clientes que geram maior receita

```python
# Clientes com o maior nmero cumulativo de pedidos (em pagamentos).
top_customers = df.groupby("customer_unique_id")["payment_value"].sum().reset_index().sort_values("payment_value", ascending=False)
top_customers.rename(columns={"payment_value":"total_paid"}, inplace=True)
```

```python
# Calcular as colunas com "% of total Sales" e "Cum % of total Sales"

top_customers['% of Total Sales'] = (top_customers['total_paid'] / top_customers['total_paid'].sum())*100

top_customers['Cum % of Total Sales'] = top_customers['% of Total Sales'].cumsum()
```

```python
#Grfico de linhas
l = len(top_customers)
fig = px.line(top_customers, x = range(1, l + 1), y = 'Cum % of Total Sales')



# definir as etiquetas do eixo x e y e o ttulo do grfico
fig.update_layout(
    xaxis_title="Nmero de Clientes",
    yaxis_title="Total Vendas Cumulativo %",
    title="Contribuio % para as vendas por nmero de clientes"
)

# adicionar uma linha de preenchimento abaixo do grfico
fig.add_shape(
    type="rect",
    xref="x",
    yref="paper",
    x0=0,
    y0=0,
    x1=40000,
    y1=1,
    fillcolor="green",
    opacity=0.2,
    layer="below"
)

# atualizar o layout da forma para ajustar a altura do preenchimento
fig.update_shapes(dict(xref='x', yref='paper'))

# adicionar um texto explicativo na figura
fig.add_annotation(
    x=55000,
    y=75,
    text="40k clientes (+-42% do total)<br> representam +-80% das vendas",
    font=dict(
        size=14,
        color="black"
    ),
    showarrow=False,
)

# exibir a figura
fig.show()
```

```python
# renomeando a coluna 'payment_value' para 'total_paid'
top_customers.rename(columns={'payment_value':'total_paid'}, inplace=True)

# Grfico de barras do Plotly
fig = px.bar(top_customers[:10], x = 'total_paid', y = 'customer_unique_id', orientation = 'h')

fig.update_layout(title = 'top 10 Clientes por Valor total Pago',
                  xaxis_title = 'Valor total Pago',
                  yaxis_title = 'ID do Cliente nico'
)


fig.show()
```

Em quais Cidades se h a maior ocorrncia de compras?

```python
#Agrupando o df por id do comprador e contando o nmero de ordem id por estado
top_pedidos_city = df.groupby('customer_state')['order_id'].count().reset_index().sort_values('order_id', ascending= False)

#renomeando as colunas "order_id' para "count'
top_pedidos_city.rename(columns={'order_id':'count'}, inplace = True)

#Criando um grfico de barras
```

```python
fig = px.bar(top_pedidos_city[:10], x = 'customer_state', y = 'count', color = 'customer_state', orientation = 'v')

fig.update_layout(
    title = 'Top 10 estados por nmero de pedidos',
    xaxis_title = 'nmero de pedidos',
    yaxis_title = 'estado')

fig.update_traces(textposition = 'outside')

fig.show()
```

Dentro de cada estado, vamos verificar quais cidades geram maior receita

```python
top_pedidosporvalor_city= df.groupby("customer_city")["payment_value"].sum().reset_index().sort_values("payment_value", ascending=False)
top_pedidosporvalor_city["% of Total Payments"] = (top_pedidosporvalor_city["payment_value"] / top_pedidosporvalor_city["payment_value"].sum()) * 100
top_pedidosporvalor_city["Cum % of Total Payments"] = top_pedidosporvalor_city["% of Total Payments"].cumsum() 
```

```python
fig = px.bar(top_pedidosporvalor_city[:10], x = "customer_city", y =  "% of Total Payments", color = 'customer_city', orientation = 'v')

fig.update_layout(
    title = 'TOP 10 Cidades por Gerao de Receita',
    xaxis_title = 'Cidade do Cliente',
    yaxis_title = 'Porcentagem do Total de Pagamentos'
)

fig.show()
```

```python
# Vamos criar um grfico de linha 
fig = px.line(top_pedidosporvalor_city, x=range(1, len(top_pedidosporvalor_city)+1), y="Cum % of Total Payments")
# atualizar as configuraes de layout do grfico

fig.update_layout(
    title = '% de Contribuio das vendas por nmero de Cidades',
    xaxis_title = 'Nmero de Cidades',
    yaxis_title = '% de contribuio para as vendas')

# Preenchimento da rea abaixo da curva
fig.add_shape(
    type = 'rect',
    xref = 'x',
    yref = 'y',
    x0 = 0,
    y0 = 0,
    x1 = 358,
    y1 = top_pedidosporvalor_city['Cum % of Total Payments'][357],
    fillcolor = 'purple',
    opacity = 0.5,
    layer = 'below',
    line_width = 0
)

# Adicionando texto ao grfico

fig.add_annotation(
    x = 1000,
    y = 70,
    text = ' 358 cidades (+- 8,7% do total) <br>contribuem para +-80% das vendas.',
    font = dict(size = 12, color = 'black'),
    showarrow = False
)

fig.show()
```

```python
print('Nmero de cidades que contribuem com 80% das vendas totais:', len(top_pedidosporvalor_city[top_pedidosporvalor_city["Cum % of Total Payments"] <= 80]),
      "ou em in %:",
      (len(top_pedidosporvalor_city[top_pedidosporvalor_city["Cum % of Total Payments"] <= 80]) / len(top_pedidosporvalor_city)) * 100)
```

## Vamos verificar o comportamento dos pedidos ao longo do perodo.

```python
# Total de pedidos por hora e por dia da semana.
df_cols = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", 
            "order_delivered_customer_date", "order_estimated_delivery_date", "shipping_limit_date", 
            "review_creation_date", "review_answer_timestamp"]

for col in df_cols:
  df[col] = pd.to_datetime(df[col])
```

```python
#Pedidos por hora

orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
pedidos_hora = orders_df.groupby(orders_df['order_purchase_timestamp'].dt.hour)['order_id'].count().reset_index().sort_values(by="order_purchase_timestamp", ascending=False)
pedidos_hora.rename(columns={"order_id":"Total Orders", "order_purchase_timestamp": "Hour of Day"}, inplace=True)
```

```python
pedidos_hora
```

```python
fig = px.bar(pedidos_hora, x = 'Hour of Day', y = 'Total Orders', title = 'Nmero de pedidos por hora do dia')
fig.update_xaxes(title = 'Hora do dia')
fig.update_yaxes(title = 'Nmero total de pedidos')
fig.show()
```

Como podemos visualizar, os pedidos comeam a aumentar por volta das 6h da manh e atingem o pico s 16h. Alm disso, o nmero de pedidos mais frequntes concentram-se durante a tarde.

#Vamos verificar total de pedidos por dia da semana.

```python
pedidos_semana = df.groupby(df['order_purchase_timestamp'].dt.day_name())['order_id'].count().reset_index()
pedidos_semana.rename(columns ={'order_id':'Total de pedidos','order_purchase_timestamp':'Dia da semana'}, inplace = True)
pedidos_semana = pedidos_semana.sort_values(by="Total de pedidos", ascending = False)
pedidos_semana
```

```python
fig = px.bar(pedidos_semana, x='Dia da semana', y='Total de pedidos', title='Nmero de pedidos por dia da semana')
fig.update_xaxes(title='Dia da semana')
fig.update_yaxes(title='Nmero total de pedidos')
fig.show()
```

Podemos ver que os maiores nmeros de pedidos esto concentrados entre s Segundas, teras e quartas.
